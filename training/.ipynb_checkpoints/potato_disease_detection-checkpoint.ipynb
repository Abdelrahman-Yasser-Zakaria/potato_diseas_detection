```python
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import models,layers  
```

```python
dataset=tf.keras.preprocessing.image_dataset_from_directory("potato",shuffle=True,image_size=(256,256),batch_size=32)
```

```python
class_names=dataset.class_names
class_names
```

```python
len(dataset) #number of batches in dataset
```

```python
plt.figure(figsize=(9,9))
for image_batch,batch_label in dataset.take(1): # take first batch
    for i in range(8):
        ax=plt.subplot(2,4,i+1)
        plt.imshow(image_batch[i].numpy().astype("int"))
        plt.title(class_names[batch_label[i]])
        plt.axis("off")
```

## split the dataset into:
- train set: 80% of the data
- validation set: 10% of the data
- test set: 10% of the data

```python
total_samples = len(dataset)
total_samples
```

Shuffle the data before splitting it.

```python
dataset=dataset.shuffle(10000,seed=12)
```

```python
# Calculate the number of samples in each set
total_samples = len(dataset)
train_samples = int(total_samples * 0.8) ## 54
val_samples = int(total_samples * 0.1) ## 6
test_samples = total_samples - train_samples - val_samples ##8

# Split the dataset
train_dataset = dataset.take(train_samples)
val_dataset = dataset.skip(train_samples).take(val_samples)
test_dataset = dataset.skip(train_samples + val_samples)

# Print the number of samples in each set
print("Train set samples:", len(train_dataset))
print("Validation set samples:", len(val_dataset))
print("Test set samples:", len(test_dataset))
```

```python
train_dataset=train_dataset.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
val_dataset=val_dataset.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
test_dataset=test_dataset.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
```
